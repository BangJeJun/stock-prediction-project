```python
df['polarity'].unique()
```

```python
# 1-> 긍정1
# -1 -> 부정2
# 0 -> 중립0
# 2 -> 긍정1
# -2 -> 부정2

# 중립:0 / 긍정:1 / 부정:2


df['polarity'] = df['polarity'].replace(['1', '-1', '0', '-2', '2'],[1, 2, 0, 2, 1])
```

```python
# index가 0-42인 행 삭제
df = df.drop(labels=range(0,43), axis=0)
df
```

```python
#인덱스 초기화 -> 0부터 정렬
df = df.reset_index(drop=True)
del df['word']
```
